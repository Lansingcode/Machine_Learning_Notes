```python
def tree_all_output(data,n,leaf):
    #data  指的是训练生成gbdt的数据集， pd.DataFrame
    #n    指训练的棵数，int类型
    #leaf  指最大叶子节点的个数，int类型
    clf = GradientBoostingClassifier(n_estimators=n,max_leaf_nodes=leaf)
    x,y=data.iloc[:,1:6].values,data.iloc[:,-1].values
    #x,y是需要根据数据情况进行更改的
    clf = clf.fit(x,y)
    leaves = clf.apply(x)[: ,: , 0]   
    df=pd.DataFrame(leaves,index=data.id)
    #每个样本在各决策树下的叶子节点
    return df
```

# 将树的叶子节点one-hot的最佳方式

```python
#将所有棵树的叶子节点one-hot
def tree_one_hot(data,n,li):
    #n 指训练的棵数，int类型
    #data 指的是训练生成gbdt的数据集， pd.DataFrame
    #l 指最大叶子节点的个数，int类型
    clf = GradientBoostingClassifier(n_estimators=n,max_leaf_nodes=li)
    x,y=data.iloc[:,1:6].values,data.iloc[:,-1].values
    #x,y是需要根据数据情况进行更改的
    clf = clf.fit(x,y)
    leaves = clf.apply(x)[: ,: , 0]   
    df=pd.DataFrame(leaves,index=data.id)
    col=[]
    for i in list(range(0,n)):
        #i是指的具体的某棵树
        df_s=pd.get_dummies(df[i],prefix='tree'+str(i+1))
        col.append(df_s)
    df_new=pd.concat(col,axis=1)
    return df_new
```

# 叶子节点的路径输出

```python
#显示某棵树的叶子节点的路径
def tree_feature_output(t,f,features):
    #t ,int类型 ，第几棵树
    #f, int类型，该叶子节点索引。
    #features ,list类型，原数据的特征变量columns 
    sub_tree = clf.estimators_[t-1][0]
    #对应树的特征
    #feature[i]表示节点i（根节点是0）用到的分裂特征对应的索引，负数表示叶子节点，没用到特征进行分裂
    #[ 2, -2,  1, -2, -2]表示根节点（0）用到了第3个特征进行分裂（从0计数，所以‘2’表示第3个）；节点（1，3，4）为叶子节点；节点2用到了第2个特征进行分裂
    sub_tree.tree_.feature
    #对应树节点的阈值
    #负数表示该节点为叶子节点
    #[ 0.5, -2. ,  1.5, -2. , -2. ]
    sub_tree.tree_.threshold
    
    #找出该叶子节点对应的路径
    #shape = [n_samples, n_nodes]
    node_indicator = sub_tree.decision_path(x) 
    node_indicator = node_indicator.toarray()   
    #找出第一个落在该叶子节点的样本对应的路径即可
    for i in range(len(node_indicator)):
        if node_indicator[i,int(f)] == 1:
            path = node_indicator[i]
            break
    path = list(path)      
    #left_children[i]表示节点i的左孩子对应的节点编号
    #right_children[i]表示节点i的右孩子对应的节点编号
    left_children = list(sub_tree.tree_.children_left)
    right_children = list(sub_tree.tree_.children_right)    
    #找出该路径对应的节点编号
    test_node = [i for i in range(int(f)+1) if path[i] == 1]
    
    #根据特征和左右关系输出规则
    for i in range(len(test_node)-1):
        #左子树小于特征阈值
        if test_node[i+1] in left_children:
            s = str(features[sub_tree.tree_.feature[test_node[i]]])+'<='+str(sub_tree.tree_.threshold[test_node[i]])+'为True'
            print(s)            
        #右子树大于特征阈值
        elif test_node[i+1] in right_children:
            s = str(features[sub_tree.tree_.feature[test_node[i]]])+'<='+str(sub_tree.tree_.threshold[test_node[i]])+'为False'
            print(s)    
```

```python
tree_feature_output(2,6,['age','gender','city','school_star','ccx_score','edu_degree'])
#第2棵树，叶子节点为6的输出路径
```
